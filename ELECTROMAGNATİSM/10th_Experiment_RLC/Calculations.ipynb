```python
# Configurations: (P/S)_xC_YR == p: parallel, S: Series, C: capasitance, R: Resistance 
file_path = "DATA/"
configurations = ["parallel/", "series/"]

# DATA/parallel

DATA = {
    "Parallel":{
        "P_1c_220R":  pd.read_csv(file_path+configurations[0]+"1c 220r" , sep="\t", decimal='.').apply(pd.to_numeric, errors='coerce'),
        "P_22c_220R": pd.read_csv(file_path+configurations[0]+"2,2c 220r" , sep="\t", decimal='.').apply(pd.to_numeric, errors='coerce'),
        "P_22c_470R": pd.read_csv(file_path+configurations[0]+"2,2c 470r" , sep="\t", decimal='.').apply(pd.to_numeric, errors='coerce'),
        "P_22c_680R": pd.read_csv(file_path+configurations[0]+"2,2c 680r" , sep="\t", decimal='.').apply(pd.to_numeric, errors='coerce'),
        "P_47c_220R": pd.read_csv(file_path+configurations[0]+"4,7c 220r" , sep="\t", decimal='.').apply(pd.to_numeric, errors='coerce')
    },
    "Series": {
        "S_1c_220R": pd.read_csv(file_path+configurations[1]+"1c 220r" , sep="\t", decimal='.').apply(pd.to_numeric, errors='coerce'),
        "S_22c_220R": pd.read_csv(file_path+configurations[1]+"2,2c 220r" , sep="\t", decimal='.').apply(pd.to_numeric, errors='coerce'),
        "S_22c_470R": pd.read_csv(file_path+configurations[1]+"2,2c 470r" , sep="\t", decimal='.').apply(pd.to_numeric, errors='coerce'),
        "S_22c_680R": pd.read_csv(file_path+configurations[1]+"2,2c 680r" , sep="\t", decimal='.').apply(pd.to_numeric, errors='coerce'),
        "S_47c_220R": pd.read_csv(file_path+configurations[1]+"4,7c 220r" , sep="\t", decimal='.').apply(pd.to_numeric, errors='coerce')
    }
}
```

```python
output_directory = "output_plots/" 
if not os.path.exists(output_directory+"Fixed_C/"):
    os.makedirs(output_directory+"Fixed_C/")
```

```python
degree_p = 8
degree_s = 20

def plot_configurations(DATA, degree):
    for configuration, combinations in DATA.items():
        # Two figures per configuration
        #plt.figure(figsize=(9,5))
        markers = itertools.cycle(['o', '+', '*','x','s', 'D', 'v', '<', '>', 'P',  ',', '1', '2', '3'])
        linestyles = itertools.cycle(['-', '--', ':', '-.',':',])
        for combination_name, combination_data in combinations.items():
            marker = next(markers)
            linestyle = next(linestyles)
            if "220R" in combination_name:
                x = combination_data["Frequency frq"].iloc[1:].dropna()
                y = combination_data["Impedance Z"].iloc[1:].dropna()
                if configuration == "Parallel":
                    
                    coefficients = np.polyfit(x, y, degree_p)
                    poly = np.poly1d(coefficients)
                    x_sorted = np.linspace(min(x), max(x), 500)  
                    plt.plot(x_sorted, poly(x_sorted), label=f'{combination_name} (Poly Fit)',linestyle=linestyle)
                elif configuration == "Series":
                    
                    coefficients = np.polyfit(x, y, degree_s)
                    poly = np.poly1d(coefficients)
                    x_sorted = np.linspace(min(x), max(x), 500) 
                    plt.plot(x_sorted, poly(x_sorted), label=f'{combination_name} (Poly Fit)',linestyle=linestyle)
                plt.scatter(x, y, alpha=0.7, label=f'{combination_name} (Data Points)',marker=marker)
        plt.xlabel('Frequency (Hz)')
        plt.ylabel('Impedance (Ohm)')
        plt.title(f'{configuration} Configurations - R Fixed at 220 Ohms')
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.savefig(output_directory+f"Fixed_R/"+f"{configuration}.png",dpi=300 )
        plt.show()
        
        markers = itertools.cycle(['o', '+', '*','x','s', 'D', 'v', '<', '>', 'P',  ',', '1', '2', '3'])
        linestyles = itertools.cycle(['-', '--', ':', '-.', '-', '--', ':', '-.', '-'])
        #plt.figure(figsize=(16, 9))
        for combination_name, combination_data in combinations.items():
            marker = next(markers)
            linestyle = next(linestyles)
            if "22c" in combination_name:
                x = combination_data["Frequency frq"].iloc[1:].dropna()
                y = combination_data["Impedance Z"].iloc[1:].dropna()
                if configuration == "Parallel":
                    coefficients = np.polyfit(x, y, degree)
                    poly = np.poly1d(coefficients)
                    x_sorted = np.linspace(min(x), max(x), 500)  
                    plt.plot(x_sorted, poly(x_sorted), label=f'{combination_name} (Poly Fit)',linestyle=linestyle)
                elif configuration == "Series":
                    coefficients = np.polyfit(x, y, degree_s)
                    poly = np.poly1d(coefficients)
                    x_sorted = np.linspace(min(x), max(x), 500)  
                    plt.plot(x_sorted, poly(x_sorted), label=f'{combination_name} (Poly Fit)',linestyle=linestyle)
                    
                plt.scatter(x, y, alpha=0.7, label=f'{combination_name} (Data Points)',marker=marker)
        plt.xlabel('Frequency (Hz)')
        plt.ylabel('Impedance (Ohm)')
        plt.title(f'{configuration} Configurations - Capacitance Fixed at 2.2 F ')
        plt.legend()
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.savefig(output_directory+f"Fixed_C/"+f"{configuration}.png",dpi=600 )
        plt.show()


plot_configurations(DATA, degree=9)
```
