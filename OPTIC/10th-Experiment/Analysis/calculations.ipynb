```python
import pandas as pd
import numpy as np
```

```python
DATA = pd.read_csv('../DATA/data.csv')
DATA
```

```python
DATA["theta_left"] = (DATA["theta_k_left"]-DATA["theta_k_left"].shift(1)).apply(lambda x: x if x > 0 else x + 2*180)
DATA["theta_right"] = (DATA["theta_k_right"]-DATA["theta_k_right"].shift(1)).apply(lambda x: x if x > 0 else x + 2*180)
DATA["theta_average"] = (DATA["theta_right"] + DATA["theta_left"])/2 # degrees
DATA["sin_theta"] = DATA["theta_average"].apply(lambda x: np.sin(np.deg2rad(x)))
DATA
```

```python
lambda_ = 650*10**-9 # m will be considered
DATA["a"] = (DATA["m"]*lambda_/DATA["sin_theta"]) # m
DATA["N"] = 1/DATA["a"] # m^-1
```

```python
DATA
```

```python
DATA["D"] = DATA["m"]/(DATA["a"]*(DATA["theta_average"].apply(lambda x: np.cos(np.deg2rad(x))))) # m
DATA["R"] = DATA["m"] * DATA["N"]
DATA["Delta_Lambda_min"] = (lambda_)/DATA["R"]
```

```python
DATA
```

```python
#save to csv # presicion of 5 decimal places

DATA.to_csv("../DATA/processed_data.csv", index=False, float_format='%.3e')
DATA
```
